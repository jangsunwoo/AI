```python
# 파이썬내부의 시스템의 기능을 가져옵니다.
import sys
from random import randint # 난수생성을 해주는 모듈
import pygame # 파이게임 모듈을 통해 게임을 만들기 위한 모듈
from pygame.locals import QUIT, Rect, KEYDOWN
# 파이게임의 기능을 초기화합니다.
pygame.init()
# 키값에 딜레이를 주어서 초기화를 시켜줍니다.
pygame.key.set_repeat(5, 5)
# 창의 이름을 바꿔줍니다.
pygame.display.set_caption("CAVE")
# 창의 크기를 설정합니다.
SURFACE = pygame.display.set_mode((800,600))
# 프레임따라 시간값을 가져옵니다.
FPSCLOCK = pygame.time.Clock()

def main():
    # 벽이 될 변수
    walls = 80
    # 플레이어가 시작될 초기 위치값
    ship_y = 250
    # 플레이어의 이동값
    velocity = 0
    # 점수값
    score = 0
    # 경사값
    slope = randint(1,6)
    # 첫번째 인자 : 폰트이름, 두번째 인자 : 폰트 크기
    sysfont = pygame.font.SysFont(None, 36)
    # ship_image에 플레이어의 이미지를 넣어줍니다.
    ship_image = pygame.image.load("ship.png")
    # over_image에 게임종료 이미지를 넣어줍니다.
    over_image = pygame.image.load("gameover.png")
    # 구멍을 가지고 있을 리스트
    holes = []
    
    #반복문을 통해서 벽을 사각형으로 만든 뒤 리스트에 넣습니다.
    for xpos in range(walls):
        holes.append(Rect(xpos * 10, 100, 10, 400))
    # 게임이 종료를 했는지 확인하는 변수
    game_over = False
    
    while True:
        # 스페이스바가 눌렸는지 체크하는 변수
        is_space_down = False
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == KEYDOWN:
                # 어떤 키가 눌렸는지 확인합니다.
                if event.key == pygame.K_SPACE:
                    is_space_down = True
        # 게임이 종료가 안되었다면 실행합니다.
        if not game_over:
            #점수를 10점씩 올립니다.
            score += 10
            # is_space_down이 false라면 -3씩 더해주고 true라면 3을 추가합니다
            velocity += -3 if is_space_down else 3
            # 플레이어의 위치에 속도값을 더해줘서 키가 눌리지 않았을때 떨어지게 해준다.
            ship_y += velocity
            # 벽을 가지고 있는 리스트 안의 마지막 인덱스를 복사해줍니다.
            edge = holes[-1].copy()
            # 벽의 x의 위치를 y의 위치만큼 움직이게 합니다.
            temp = edge.move(0, slope) 
            if temp.top <= 0 or temp.bottom >= 600:
                slope = randint(1, 6) * (-1 if slope > 0 else 1)
                #inflat_ip? 사각형을 늘리거나 줄이는 메서드
                edge.inflate_ip(0, -20)
            edge.move_ip(10, slope)
            #edge에는 사각형이 그려져 있고, 경사도, 움직임값 저장, 
            #리스트의 맨 뒷 인덱스에 추가
            holes.append(edge)
            # walls -> 80, holes-> walls의 길이, 80개를 유지해줘야하기 때문에
            del holes[0]
            # holes의 기능들을 x안에다가 넣어주고 x를 x축으로 이동시킵니다.
            holes = [x.move(-10,0) for x in holes]
            
            if holes[0].top > ship_y or holes[0].bottom < ship_y + 80 :
                game_over = True
            
            SURFACE.fill((0,255,0))
            for hole in holes:
                pygame.draw.rect(SURFACE, (0,0,0), hole)
            # 플레이어 이미지를 배치합니다.
            SURFACE.blit(ship_image, (0, ship_y))
            score_image = sysfont.render("score : {}".format(score), True, (0,0,255))
            SURFACE.blit(score_image, (600, 20))
            
            if game_over:
                SURFACE.blit(over_image, (280,180))
                
            pygame.display.update()
            FPSCLOCK.tick(15)
if __name__ == '__main__':
    main()
            
            
```

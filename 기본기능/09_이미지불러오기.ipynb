```python
import cv2

# 이미지를 읽습니다.
image= cv2.imread("Steve.jfif", 0)

cv2.imshow("SteveJobs", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 다른 이름으로 저장
cv2.imwrite("SteveJobs.png", image)
```

```python
#video = cv2.VideoCapture(0) # 웹캡을 사용할때 사용합니다.
video = cv2.VideoCapture("Changes.webm")

while video.isOpened():
    success, frame = video.read()
    if success:
        cv2.imshow("Video", frame)
        
        # ESC를 누르면 종료
        #waitKey(time) => time마다 프레임을 재생합니다.
        key = cv2.waitKey(1) & 0xFF
        if key == 27:
            break
video.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
# 선형타입 종류
# FILLED = 내부 채우기
# LINE_4 = 4 점 이웃 연결
# LINE_8 = 8 점 이웃 연결
# LINE_AA = 안티앨리어싱
src = np.zeros((768, 1366, 3), dtype=np.uint8)
# line(이미지, (시작좌표(x1,y1)), (끝나는좌표(x2,y2), 색상값(BGR), 선두께, 선형타입))
cv2.line(src,(100,100),(1200,100),(0,255,0),2, cv2.LINE_AA)
#circle(이미지, 그릴좌표, 반지름, (색상값), 두께, 선형타입)
# 원안을 다 채우고싶다면 두께의 위치 cv2.FILLED
cv2.circle(src, (100,250), 50, (255,0,0), 1, cv2.LINE_AA)
# rectangle(이미지, 시작할좌표, 끝날좌표, 색상값, 두께, 선형타입)
cv2.rectangle(src, (300,250), (700, 400),(255,255,0), cv2.FILLED, cv2.LINE_8)
# ellipse(이미지, 시작할좌표, (중심이 가장 먼 거리, 중심에서 가장 가까운 거리),
# 각도, 시작 각도, 종료 각도, 색상값, 두께, 선형타입)
cv2.ellipse(src, (1000, 300), (100, 80), 0, 90, 180, (255, 0, 255), 2)
# putText(이미지, 쓸 문자, (시작좌표), 글꼴, 글자크기, 색상값, 두께, 선형타입)
cv2.putText(src, "PYTHON ROCK", (300, 500), cv2.FONT_HERSHEY_COMPLEX, 2, (255,255,255), 3)
cv2.imshow("Figures", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

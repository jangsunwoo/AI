```python
from bs4 import BeautifulSoup
html = "<html> \
            <body> \
                <h1 id='title'>[1]크롤링이란?</h1> \
                <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p> \
                <p id = 'body' align = 'center'>파이썬을 중심으로 다양한 웹크롤링 기술이 발달되어 있습니다.</n = 'center'>p> \
            </body> \
        </html>"
soup = BeautifulSoup(html, 'html.parser')
data = soup.find('p', class_='cssstyle')
print(data.get_text())
data2 = soup.find('p', attrs={'align':'center'})
print(data2.string)
alldata = soup.find_all('p', 'span')
print(alldata)
for item in alldata:
    print(item.get_text())
```

# selector
* css selector 기능으로 html 데이터를 선택할때
    * tag의 이름으로 선택 : select('a')
    * tag + class의 이름으로 선택 가능 합니다. : select('span.ah_k')
    * tag + 여러가지 클래스 이름으로 선택 가능
    * 클래스의 이름으로만 선택 가능 :

```python
import requests
from bs4 import BeautifulSoup
site_list = ["https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000003", "https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000001", "https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000"]
for site in site_list:
    # 사이트를 가지고 있을 res
    res = requests.get(site)
    # 사이트가 가진 컨텐츠를 html.parser의 기능을 통해 가져오고
    soup=  BeautifulSoup(res.content, 'html.parser')
    # select-> 기능중에 #productListArea->ul->li->p->a 컨텐츠부분을 다 가져옵니다.
    data = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for item in data:
        # data에 있던 컨텐츠를 다 나눠서 텍스트만 출력합니다.
        print(item.get_text())
```

```python
#-*- coding:utf-8 -*-
import sys
import requests
from bs4 import BeautifulSoup
res = requests.get('https://tv.naver.com/r')
raw = res.text
soup=  BeautifulSoup(raw, 'html.parser')
# html안에 어디가에 있는 태그의 이름을 찾아서 내용을 가지고 있게 할 수도 있다.
data = soup.select('div.cds')
text = open('Top100.txt','w', encoding='utf-8')
for item in data:
    title = item.select_one('tooltip').text
    hit = item.select_one("span.hit").text
    text.write(title)
    text.write(" / ")
    text.write(hit)
    text.write("\n")
    
```
